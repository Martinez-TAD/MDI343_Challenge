```python
import pandas as pd
#import seaborn as sns
#import matplotlib.pyplot as plt
import sklearn as sk
import numpy as np

import lightgbm
```

```python
xtrain=pd.read_csv("data/xtrain_challenge.csv")
ytrain=pd.read_csv("data/ytrain_challenge.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Prenons juste les valeurs S1 Ã  S11
cols_S1_11 = list(xtrain.columns)[26:]
```

```python
X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(xtrain, ytrain, test_size=0.33, random_state=42)
```

```python
X_train_s = X_train[cols_S1_11]/10000.
```

```python
X_train_s.head(10)
```

```python
 # making lgbm datasets for train and valid
d_train = lightgbm.Dataset(X_train, y_train)
d_valid = lightgbm.Dataset(X_test, y_test)
    
parameters = {
    'application': 'binary',
    'objective': 'binary',
    'metric': 'binary_error',
    'is_unbalance': 'false',
    'boosting': 'gbdt',
    'num_leaves': 100,
    'feature_fraction': 0.6,
    'bagging_fraction': 0.8,
    'bagging_freq': 20,
    'learning_rate': 0.04,
    'verbose': 0
}

model = lightgbm.train(parameters,
                       d_train,
                       valid_sets=d_valid,
                       num_boost_round=5000,
                       early_stopping_rounds=150)   
    # making prediciton for one column
```

```python

soft = model.predict(X_test)

print('Fininshed Training')
soft
```

```python

yvalid = soft
```

```python
N = np.sum(y_test == 0).values
P = np.sum(y_test == 1).values

```

```python

yvalid_scoreordered = y_test.values[np.argsort(yvalid)]


FP = 0.0
TP = 0.0
val = 0
print(N)
print(P)
for i in range(len(y_test) - 1, -1, -1):        
    if (yvalid_scoreordered[i] == 1):
        TP = TP + 1
    else:
        FP = FP + 1
        #print(X_test[cols].iloc(i))
    if (FP / N > 10**-4):
        FP = FP - 1
        break
print("For the smallest FPR <= 10^-4 (i.e., ", FP / N, ") TPR = ", TP / P, ".", sep = "")
```
