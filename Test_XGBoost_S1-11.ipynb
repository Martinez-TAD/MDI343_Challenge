```python
xtrain=pd.read_csv("data/xtrain_challenge.csv")
ytrain=pd.read_csv("data/ytrain_challenge.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
print(f"Train class 1: {ytrain.sum().values[0]}")
print(f"Train class 0: {(ytrain.count() - ytrain.sum()).values[0]}")

# Classes tres desiquilibrées

scale_pos_weight = np.sqrt((ytrain.count() - ytrain.sum()).values[0]/ytrain.sum().values[0]) 
# Prenons juste les valeurs S1 à S11
cols_q1_16 = list(xtrain.columns)[0:26]
cols_S1_11 = list(xtrain.columns)[26:]
```

```python
X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(xtrain, ytrain, test_size=0.33, random_state=42)
```

```python
X_train_s = X_train[cols_S1_11]
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(scale_pos_weight=scale_pos_weight, learning_rate=0.8, n_estimators=50 , n_jobs=-1)
%time xgb.fit( X_train_s.values, y_train.values)
```

```python

soft = xgb.predict_proba(X_test[cols_S1_11].values)[:,1]
```

```python
print(soft)
yvalid=soft
```

```python

yvalid_scoreordered = y_test.values[np.argsort(yvalid)]

N = np.sum(y_test == 0).values
P = np.sum(y_test == 1).values

FP = 0.0
TP = 0.0
val = 0
print(N)
print(P)
for i in range(len(y_test) - 1, -1, -1):        
    if (yvalid_scoreordered[i] == 1):
        TP = TP + 1
    else:
        FP = FP + 1
        #print(X_test[cols].iloc(i))
    if (FP / N > 10**-4):
        FP = FP - 1
        break
print("For the smallest FPR <= 10^-4 (i.e., ", FP / N, ") TPR = ", TP / P, ".", sep = "")
```

```python
# Load test data
xtest = pd.read_csv('data/xtest_challenge.csv')
```

```python
# Classify the provided test data
ytest = xgb.predict_proba(xtest[cols_S1_11].values/10000.)[:,1]
print(ytest.shape)
np.savetxt('ytest_xgboost_1.csv', ytest, fmt = '%1.15f', delimiter=',')
```
